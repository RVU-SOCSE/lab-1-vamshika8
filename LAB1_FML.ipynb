```python
#LAB1
#Bias–Variance Trade-off
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Study hours (x)
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
# Marks (y)
y = np.array([52, 55, 61, 70, 82])
```

```python
# Create linear regression model
model_A = LinearRegression()
# Train (fit) the model
model_A.fit(x, y)
# Get coefficients
beta_0 = model_A.intercept_
beta_1 = model_A.coef_[0]
print("Model A Equation:")
print(f"y = {beta_0:.2f} + {beta_1:.2f}x")
```

```python
# Convert x into polynomial features (x, x^2, x^3, x^4)
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(x)
# Create linear regression model for polynomial data
model_B = LinearRegression()
model_B.fit(x_poly, y)
# Get coefficients
coefficients = model_B.coef_
intercept = model_B.intercept_
print("\nModel B Equation:")
print("y =", intercept)
for i in range(1, len(coefficients)):
    print(f"+ ({coefficients[i]:.4f})x^{i}")
```

```python

# Predict using Model A
x_new = np.array([[6]])
y_pred_A = model_A.predict(x_new)
# Predict using Model B
x_new_poly = poly.transform(x_new)
y_pred_B = model_B.predict(x_new_poly)
print("\nPrediction when x = 6 hours:")
print("Model A Prediction:", y_pred_A[0])
print("Model B Prediction:", y_pred_B[0])
```

```python

# Predictions on training data
y_train_pred_A = model_A.predict(x)
y_train_pred_B = model_B.predict(x_poly)
# Mean Squared Error
mse_A = mean_squared_error(y, y_train_pred_A)
mse_B = mean_squared_error(y, y_train_pred_B)
print("\nMean Squared Error:")
print("Model A MSE:", mse_A)
print("Model B MSE:", mse_B)
print("\nBias–Variance Analysis:")
print("Model A has HIGH BIAS and LOW VARIANCE (Underfitting)")
print("Model B has LOW BIAS and HIGH VARIANCE (Overfitting)")
```
